```python
import geopandas as gpd
import pandas as pd
import numpy as np
```

### Read file

```python
!ls
```

```python
# Get poverty.zip from: https://geoportal.dane.gov.co/visipm/
df = gpd.read_file('Tabular data/poverty.zip')
df
```

```python
df = df[['COD_MPIO', 'ipm', 'LABEL']]

df = df[~df.LABEL.isna()]

df
```

```python
poverty = df.groupby('COD_MPIO').agg({'ipm': 'mean'}).reset_index()
poverty.rename(columns={'ipm': 'poverty', 'COD_MPIO': 'Municipality code'}, inplace=True)
poverty['Municipality code'] = poverty['Municipality code'].astype(int)
poverty
```

```python
dengue = pd.read_csv('Tabular data/dengue_tabular.csv')
dengue
```

## Merge files

```python
dengue = pd.merge(dengue, poverty, on='Municipality code', how='left')
dengue
```

```python
dengue.to_csv('dengue_tabular_v2.csv', index=False)
```
